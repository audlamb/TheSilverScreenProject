```python
import pandas as pd
import datetime
import numpy as np
```

```python
date_data = '../TheMoviedb_query/MovieQuery2_Clean'
date_df = pd.read_csv(date_data, parse_dates=True)
date_df.head()
```

```python
date_df['Award_date'] = ""
date_df.head()
```

```python
date_table=date_df[['Movie','Release Date','Award_date']]
date_table.head()
```

```python
date_table['Release_Year'] = pd.DatetimeIndex(date_table['Release Date']).year
date_table.head()
```

```python
bins = [0, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]
labels = ['2005-02-27', '2006-03-05', '2007-02-25', '2008-02-24', '2009-02-22', '2010-03-07',
          '2011-02-27', '2012-02-26', '2013-02-24', '2014-03-02', '2015-02-22', '2016-02-28']

date_table['Award_date'] = pd.cut(date_table['Release_Year'], bins=bins, labels=labels)
date_table
```

```python
date_table = date_table.replace('', np.nan)
date_table = date_table.dropna()
```

```python
date_table['Days_Difference'] = date_table['Release Date'].sub(date_table['Award_date'], axis=0)
date_table.head()         
```
