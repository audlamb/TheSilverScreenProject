```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

```python
csv_file = 'TheSilverScreenProject/OMDB_TMDB_Merge/AllData_withOscarWins.csv'

data_df = pd.read_csv(csv_file)
data_df.head(2)
```

```python
new_index_df = data_df.set_index("Win or Nominated")
nomination_df = new_index_df.loc['N']
nomination_df.head(1)
```

```python
oscar_df = new_index_df.loc['W']
oscar_df.head(1)
```

```python
# Build the scatter plots

fig, ax = plt.subplots(sharex=True)

nomination_df.plot( kind='scatter', subplots=True, x='Metascore', y='Oscar Wins',
                   c='blue', ax=ax, alpha=0.5, edgecolors='black', label='Nominated')
oscar_df.plot( kind='scatter', subplots=True, x='Metascore', y='Oscar Wins',
                   c='red', ax=ax, alpha=0.5, edgecolors='black', label='Wins')

# Incorporate the other graph properties

ax.grid()
ax.set_axisbelow(False)
plt.xlabel("Metascore")
plt.ylabel("Number of Oscars or Nominations")
plt.title("Metascore vs. Nominations and Wins")

# Create a legend

lgnd = plt.legend(loc="best", title="Type", fontsize="small")

# Save Figure

# plt.savefig('Metascore.png')
```
