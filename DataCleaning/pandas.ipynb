```python
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/database.csv"

# Read Purchasing File and store into Pandas data frame
movie_data = pd.read_csv(file_to_load)
movie_data.head()
```

```python
movie_data = movie_data.loc[movie_data['Ceremony']> 77]
movie_data.head()
```

```python
movie_table=pd.DataFrame()
movie_table=movie_data[['Award','Winner','Film']]
movie_table.to_csv("wholelist.csv", encoding="utf-8", index=False)
movie_table.head()
```

```python
movie_winner = movie_table.loc[movie_table['Winner'] == 1.0]
movie_winner.to_csv("winner.csv", encoding="utf-8", index=False)
movie_winner.head()
```

```python
movie_loser = movie_table.loc[movie_table['Winner'] != 1.0]
movie_loser.to_csv("loser.csv", encoding="utf-8", index=False)
movie_loser.head()
```
