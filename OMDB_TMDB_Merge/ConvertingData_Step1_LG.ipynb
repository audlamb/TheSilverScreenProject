```python
import pandas as pd
import csv
```

```python
finalm_data = 'TheSilverScreenProject/OMDB_TMDB_Merge/parsed_movies.csv'

finalm_df = pd.read_csv(finalm_data)
finalm_df = finalm_df[['Movie', 'Release Date_x','Release Date_y', 'Budget', 'Revenue', 'Box Office', 'Genres', 'Rating', 'Runtime', 'Metascore', 'IMDb Rating', 'Awards', 'Production Company', 'wins', 'nominations']]
finalm_df.head()
```

```python
#finalm_df['B'] = finalm_df['Awards'].str.extract("(\d*\.?\d+)", expand=True)
#finalm_df.head(200)
```

```python
finalm_df['Awards'] = finalm_df['Awards'].fillna('x1 x2 x3 x4')
finalm_df['Awards'] = finalm_df['Awards'].str.replace('Nominated for ', 'N')
finalm_df['Awards'] = finalm_df['Awards'].str.replace('Won ', 'W')
finalm_df['Awards'] = finalm_df['Awards'].str.replace(' wins', 'wins')
finalm_df['Awards'] = finalm_df['Awards'].str.replace(' Oscars', 'O')
finalm_df['Awards'] = finalm_df['Awards'].str.replace(' Oscar', 'O')
finalm_df['Awards'] = finalm_df['Awards'].str.replace('Another ', 'A')
finalm_df['Awards'] = finalm_df['Awards'].str.replace(' nominations', 'nom')
finalm_df['Awards'] = finalm_df['Awards'].str.replace(' Golden Globe', 'GG')
finalm_df['Awards'] = finalm_df['Awards'].str.replace(' nomination', 'nom')
finalm_df['Awards'] = finalm_df['Awards'].str.replace(' win', 'wins')
finalm_df['Awards'] = finalm_df['Awards'].str.replace(' & ', '')
finalm_df['Awards'] = finalm_df['Awards'].str.replace('1wins ', 'x1 1wins ')
finalm_df['Awards'] = finalm_df['Awards'].str.replace('A', '++')
finalm_df['Awards'] = finalm_df['Awards'].str.replace('nom.', '')
finalm_df['Awards'] = finalm_df['Awards'].str.replace('wins', ',')
finalm_df['Awards'] = finalm_df['Awards'].str.replace('.', ',')
finalm_df['Awards'] = finalm_df['Awards'].str.replace('x1', '0,')
finalm_df['Awards'] = finalm_df['Awards'].str.replace('x2', '0,')
finalm_df['Awards'] = finalm_df['Awards'].str.replace('x3', '0,')
finalm_df['Awards'] = finalm_df['Awards'].str.replace('x4', '0,')
finalm_df['Awards'] = finalm_df['Awards'].str.replace('1 ', '++1')
finalm_df['Awards'] = finalm_df['Awards'].str.replace('1,', '++1,')
finalm_df['Awards'] = finalm_df['Awards'].astype(str)
```

```python
#finalm_df['B'] = finalm_df['Awards'].str.extract("(\d*\.?\d+)", expand=True)
finalm_df[['Win or Nominated','Oscar Wins']] = finalm_df['Awards'].str.extract(r'([WN])?(\d)')
finalm_df.head(100)
```

```python
finalm_df.to_csv('AllData_withOscarWins.csv')
```

```python
finalm_df['Production Company'] = finalm_df['Production Company'].fillna('Not Listed')
finalm_df['Release Date_x'] = finalm_df['Release Date_x'].fillna('Not Listed')
finalm_df['Release Date_y'] = finalm_df['Release Date_y'].fillna('Not Listed')
finalm_df['Budget'] = finalm_df['Budget'].fillna('0')
finalm_df['Revenue'] = finalm_df['Revenue'].fillna('0')
finalm_df['Box Office'] = finalm_df['Box Office'].fillna('0')
finalm_df['Genres'] = finalm_df['Genres'].fillna('Not Listed')
finalm_df['Rating'] = finalm_df['Rating'].fillna('Not Listed')
finalm_df['Runtime'] = finalm_df['Runtime'].fillna('0')
finalm_df['Metascore'] = finalm_df['Metascore'].fillna('0')
finalm_df['IMDb Rating'] = finalm_df['IMDb Rating'].fillna('0')
finalm_df['wins'] = finalm_df['wins'].fillna('0')
finalm_df['nominations'] = finalm_df['nominations'].fillna('Not Listed')
finalm_df.count()
```

```python
finalm_df = finalm_df.dropna(how = 'any')
finalm_df.count()
```
