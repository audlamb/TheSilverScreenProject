```python
import pandas as pd
import csv
```

```python
themoviedb_data = 'TheSilverScreenProject/TheMoviedb_query/Mergedquery_clean.csv'
omdb_data = 'TheSilverScreenProject/OMDb_Data/OMDb_Query.csv'

tmdb_df = pd.read_csv(themoviedb_data)
omdb_df = pd.read_csv(omdb_data)
```

```python
tmdb_df.head(5)
```

```python
omdb_df.head(5)
```

```python
fullquery_df = pd.merge(tmdb_df, omdb_df, on = 'Movie', how = 'right')
fullquery_df.head(2)
```

```python
fullquery_df = fullquery_df[['Movie','Release Date_x','Release Date_y', 'Budget', 'Revenue', 'Box Office', 'Genres', 'Rating', 'Runtime', 'Metascore', 'IMDb Rating', 'Awards', 'Production Company']]
fullquery_df.head()
```

```python
fullquery_df.count()
```

```python
fullquery_df.describe()
```

```python
fullquery_df.dtypes
```

```python
fullquery_df['Box Office'] = fullquery_df['Box Office'].str.replace(',','')
fullquery_df['Box Office'] = fullquery_df['Box Office'].str.replace('$', '')
fullquery_df['Box Office'] = fullquery_df['Box Office'].fillna(0)
fullquery_df['Box Office'] = fullquery_df['Box Office'].astype(float)
```

```python
fullquery_df.head(100)
```

```python
fullquery_df.dtypes
```

```python
fullquery_df['Runtime'] = fullquery_df['Runtime'].str.replace(' min','')
fullquery_df['Runtime'] = fullquery_df['Runtime'].fillna(0)
fullquery_df['Runtime'] = fullquery_df['Runtime'].astype(float)
fullquery_df.head()
```

```python
fullquery_df.dtypes
```

```python
fullquery_df.to_csv('OMDB_TMDB_DataMerged.csv')
```
