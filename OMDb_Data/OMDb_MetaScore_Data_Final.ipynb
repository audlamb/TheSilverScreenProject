```python
import requests 
import pandas as pd
import json
import time
from pprint import pprint
```

```python
merged_movie_data = 'TheSilverScreenProject/TheMoviedb_query/Mergedquery_clean.csv'

list_df = pd.read_csv(merged_movie_data)
list_df.head(5)
#purchase_df.count()
```

```python
dflist = list_df['Movie'].tolist()
#print(len(dflist))
```

```python
url = "http://www.omdbapi.com/?apikey=trilogy&t="
api_key = "59ef3a43"
```

```python
metascore_df = pd.DataFrame(columns = ['Movie', 
                            'Rating',
                            'Release Date',
                            'Runtime',
                            'Metascore',
                            'IMDb Rating',
                            'Box Office',
                            'Awards',
                            'Production Company'])
metascore_df.head()
```

```python
call_count = 1
error_count = 0
index = 0

for movie in dflist:
    
    try:
        print(f'Processing # {call_count} | {movie}')
        movie_data = requests.get(url + movie).json()
        metascore_df.set_value(index, "Movie", movie_data["Title"])
        metascore_df.set_value(index, "Rating", movie_data["Rated"])
        metascore_df.set_value(index, "Runtime", movie_data["Runtime"])
        metascore_df.set_value(index, "Release Date", movie_data["Released"])
        metascore_df.set_value(index, "Metascore", movie_data["Metascore"])
        metascore_df.set_value(index, "IMDb Rating", movie_data["imdbRating"])
        metascore_df.set_value(index, "Box Office", movie_data["BoxOffice"])
        metascore_df.set_value(index, "Awards", movie_data["Awards"])
        metascore_df.set_value(index, "Production Company", movie_data["Production"]) 
        index = index + 1
    except Exception as error:
        print(type(error))
        print(str(error))
        error_count = error_count + 1 
    call_count = call_count + 1 
        
print(f'--------------- The Search is Over! ---------------')
print(error_count)        
        
        
```
